```python
def pad_to_center(image, out_size=(800,1280)):
    out_image = np.zeros(out_size)
    delta_h = out_size[0]-image.shape[0]
    delta_w = out_size[1]-image.shape[1]
    out_image[delta_h//2:-delta_h//2,delta_w//2:-delta_w//2] = image
    return out_image
```
